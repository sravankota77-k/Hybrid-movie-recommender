```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
movies
```

```python
print(movies.columns)
```

```python
movies[['title','genres','keywords','overview']].head()
```

```python
import ast
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return " ".join(L)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies[['title','genres','keywords']].head()
```

```python
movies['combined_features'] = (
    movies['genres'] + " " +
    movies['keywords'] + " " +
    movies['overview']
)
```

```python
movies['combined_features'].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
movies['combined_features'] = movies['combined_features'].fillna('')
feature_vectors = vectorizer.fit_transform(
    movies['combined_features']
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(feature_vectors)
```

```python
# rating
movies[['title','vote_average','vote_count']].head()
```

```python
C = movies['vote_average'].mean()
m = movies['vote_count'].quantile(0.90)
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m)*R) + (m/(m+v)*C)
```

```python
movies['score'] = movies.apply(weighted_rating, axis=1)
```

```python
def recommend(movie_name):

    index = movies[movies['title']==movie_name].index[0]

    distances = list(enumerate(similarity[index]))

    similar_movies = sorted(
        distances,
        key=lambda x:x[1],
        reverse=True
    )[1:20]

    recommendations = []

    for i in similar_movies:
        recommendations.append(
            movies.iloc[i[0]]
        )

    rec_df = pd.DataFrame(recommendations)

    rec_df = rec_df.sort_values(
        by='score',
        ascending=False
    )

    for movie in rec_df['title'].head(5):
      print(movie)
```

```python
recommend("Avatar")
```

```python
recommend("Batman Begins")
```
